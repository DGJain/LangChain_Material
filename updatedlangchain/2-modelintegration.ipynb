### Streaming and Batch 

Streaming 

most models can stream their output content while it is being generated. By displaying output progressively , streaming significantly improves user experience , particularly for longer responses. Calling stream() returns an iterator that yields chunks as they are produced. You can use a loop to process each chunk in real-time 

```python
model.invoke("write me a 200 words paragraph on artificial intelligence")

```

```python
model.stream("write me a 200 words paragraph on artificial intelligence")
```

```python
for chunk in model.stream("write me a 200 words paragraph on artificial intelligence"):
    print(chunk.text, end="|" ,flush=True)
    
```

```python
from langchain_groq import ChatGroq
model = ChatGroq(model="qwen/qwen3-32b")
for chunk in model.stream("why do parrots have colourful feathers"):
    print(chunk.text, end="" ,flush=True)
```

### Batch

Batching a collection of inpendent requests to a model can significantly improve performance and reduce costs , as the processing can be done in parallel

```python
responses = model.batch([
    "why do parrots have colourful feathers?",
    "how do airplanes fly",
    "what is quantum computing"
    ])

for response in responses:
    print(response)
```

```python
model.batch([
    "why do parrots have colourful feathers?",
    "how do airplanes fly",
    "what is quantum computing",
    "who is the pm of india",
    "who is the richest man in 2026",
    "who won 2022 fifa world cup"],
    config={
        'max_concurrency': 5, # limit to 5 parallel cells
    }
    )
```
